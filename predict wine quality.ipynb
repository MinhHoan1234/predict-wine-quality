import sklearn


```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

#### Get the data test

```python
df1 = pd.read_csv("C:\\Users\\Admin\\OneDrive\\Máy tính\\tài liệu học tập\\AIL303m\\kaggle\\test.csv")
print(f"sample size = {df1.shape[0]}\nnumber of columns = {df1.shape[1]}")
df1.head()
```

```python
df1['type'] = df['type'].replace({'white':0,'red':1})
df1
```

```python
new_df = df.drop(df.columns[-2],axis=1)
new_df.head()
```

```python
X_train = new_df.iloc[:,:].values # all cols and rows except the last(quality)
Y_train = df.iloc[:,-2].values #the col quality
X_test = df1.iloc[:,1:-1].values # all except type
Y_test = df1.iloc[:,-1].values
```

```python
print('X_train shape:',X_train.shape)
print('Y_train shape:',Y_train.shape)
print('X_test shape:',X_test.shape)
print('Y_test shape:',Y_test.shape)
```

```python
X_test
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

#### Random Forest Classifier

```python
# Create Random Forest Classifier Instance

rf_regressor = RandomForestClassifier(n_estimators = 100)

# Fitting the classifier with data training
rf_regressor.fit(X_train,Y_train)

# Predicting using Classifier
Y_pred1 = rf_regressor.predict(X_test)
```

```python
Y_pred1
```

```python
for a in Y_pred1:
    print(a)
```

#### Support vector classifier

```python
svc_regression = SVC()

# Fitting train data
svc_regression.fit(X_train,Y_train)

#predict
Y_pred2 = svc_regression.predict(X_test) 
```

```python
print(Y_pred2)
```

```python
for a in Y_pred2:
    print(a)
```

#### Tuning ML algorithm

```python
# Find parameters for SVC model
param = {
    'C': [0.1,0.8,1,1.1,1.2,1.3,1.4],
    'kernel': ['linear','rbf'],
    'gamma': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]
}
grid_svc = GridSearchCV(svc_regression, param_grid=param,scoring='accuracy',cv = 10)
```

```python
grid_svc.fit(X_train,Y_train)
```

```python
print("Best parameters: ",grid_svc.best_params_)
```

```python
svc_regressor2 = SVC(C = 1.4, gamma= 1.4, kernel= 'rbf')
svc_regressor2.fit(X_train, Y_train)
y_pred_svc = svc_regressor2.predict(X_test)
```

```python
for a in y_pred_svc:
    print(a)
```
